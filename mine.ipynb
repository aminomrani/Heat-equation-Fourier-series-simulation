```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1
T0 = 100
alpha = 0.01
terms = 100

def T_xt(x, t, L, T0, alpha, terms):
    result = np.zeros_like(x)
    for n in range(1, terms + 1, 2):
        Bn = (4 * T0) / (n * np.pi)
        term = Bn * np.sin(n * np.pi * x / L) * np.exp(-alpha * (n * np.pi / L)**2 * t)
        result += term
    return result
x = np.linspace(0, L, 200)
times = np.arange(0, 2,0.1 )
plt.figure(figsize=(10, 6))
for t in times:
    T = T_xt(x, t, L, T0, alpha, terms)
    plt.plot(x, T, label=f"t = {t:.2f} s")

plt.title("Temperature Distribution T(x, t) in the Rod Over Time")
plt.xlabel("Position along the rod (x)")
plt.ylabel("Temperature T(x, t)")
plt.grid(True)
plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1.0))
plt.tight_layout()
plt.show()
```
